```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("Meteorite_Landings (6).csv")

# عرض أول 5 صفوف
df.head()
top_heavy = df.sort_values(by='mass (g)',ascending=False)
top_heavy[['name','mass (g)','year','fall']].head()
df['year'].value_counts().head(10)
plt.figure(figsize=(14,7))
sns.scatterplot(
    x = "reclong",
    y = "reclat",
    alpha = 0.9,
    s = 8,
    color = 'red',
    data = df
)
plt.text(-100, 40, 'North America', fontsize=12, color='black', fontweight='bold', bbox=dict(facecolor='white', alpha=0.6))
plt.text(-60, -15, 'South America', fontsize=12, color='black', fontweight='bold', bbox=dict(facecolor='white', alpha=0.6))
plt.text(15, 15, 'Africa', fontsize=12, color='black', fontweight='bold', bbox=dict(facecolor='white', alpha=0.6))
plt.text(20, 50, 'Europe', fontsize=12, color='black', fontweight='bold', bbox=dict(facecolor='white', alpha=0.6))
plt.text(90, 40, 'Asia', fontsize=12, color='black', fontweight='bold', bbox=dict(facecolor='white', alpha=0.6))
plt.text(135, -25, 'Australia', fontsize=12, color='black', fontweight='bold', bbox=dict(facecolor='white', alpha=0.6))
plt.title('Global Meteorite Landings: Identifying High & Low Density Impact Zones', fontsize=20, fontweight='black')
plt.xlabel('Longitude (Darker clusters indicate higher density of found meteorites)', fontsize = 20)
plt.ylabel('Latitude', fontsize=20)
plt.show()
df.isnull().sum()
df_cleaned = df.dropna(subset=['reclong','reclat'])## دالة تنظيف ف python
df_cleaned.isnull().sum()
df_cleaned.loc[:,'mass (g)'] = df_cleaned['mass (g)'].fillna(0)
df_cleaned.loc[:,'year'] = df_cleaned['year'].fillna(0)
df_cleaned.isnull().sum()
##########################
fall_counts = df_cleaned['fall'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(
    fall_counts, 
    labels=fall_counts.index, 
    autopct='%1.1f%%',       # لإظهار النسبة المئوية على الرسمة
    startangle=140,          # زاوية بدء الرسم لشكل متناسق
    colors=['skyblue', 'salmon']
)

plt.title('Meteorite Status: Found vs. Fell', fontsize=14, fontweight='bold')

plt.show()
```
